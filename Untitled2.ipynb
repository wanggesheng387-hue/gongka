```python
# @title
import requests
import json

def test_ollma_model(prompt):
    url = "http://localhost:11434/api/generate"
    data = {
        "model": "llama3",  # 已下载的模型名
        "prompt": prompt,
        "stream": False
    }
    response = requests.post(url, json=data)
    result = json.loads(response.text)
    return result.get("response", "")

# 测试用例
if __name__ == "__main__":
    test_prompt = "请简要介绍Ollma"
    output = test_ollma_model(test_prompt)
    print("模型输出：", output)
    # 可以添加断言判断结果是否符合预期
    assert len(output) > 0, "测试失败：未获取到模型输出"
    print("测试通过！")
```
